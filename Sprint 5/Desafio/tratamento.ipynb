<h2>Criando o DataFrame</h2>

```python
import pandas as pd

dados = pd.read_csv(r"C:\Users\arthu\OneDrive\Desktop\Programa de Bolsas\PB-Compass\Sprint 5\Desafio\PALAS_OPERACOES_2023_07.csv", delimiter=";")

dados
```

<h2>Percebi que algumas colunas do DataFrame possuem alguns espaços desnecessários, então decidi usar o str.strip() para tira-los e além disso, substitui todos os vazios e nulos por 0</h2>

```python
dados["Area"] = dados["Area"].str.strip()
dados["Sigla Unidade Institucional"] = dados["Sigla Unidade Institucional"].str.strip()
dados["Atuacao em Territorio de Fronteira"] = dados["Atuacao em Territorio de Fronteira"].str.strip()


dados.replace("", "0", inplace=True)
dados.fillna(0, inplace=True)

dados
```

- No tratamento acima, percebi que na coluna "Atuacao em território de Fronteira", um "Nao" se encaixaria melhor no lugar do 0, então fiz essa substituição:

```python
dados["Atuacao em Territorio de Fronteira"] = dados["Atuacao em Territorio de Fronteira"].astype(str)

dados["Atuacao em Territorio de Fronteira"].replace("0", "Nao", inplace=True)

dados
```

<h2>Exportando DataFrame tratado</h2>

```python
dados.to_csv("PALAS_OPERACOES_2023_07-TRATADO", index=False, encoding='utf-8')
```
